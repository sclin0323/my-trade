```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../etf_data/AADR.csv')
```

```python
close = list(reversed(df['Adj Close']))
```

```python
ma10 = []
ma20 = []
ma60 = []
ma120 = []
ma250 = []
for i in range(len(close)-9):
    ma10.append(np.average(close[i:i+10]))
for i in range(len(close)-19):
    ma20.append(np.average(close[i:i+20]))
for i in range(len(close)-59):
    ma60.append(np.average(close[i:i+60]))
for i in range(len(close)-119):
    ma120.append(np.average(close[i:i+120]))
for i in range(len(close)-249):
    ma250.append(np.average(close[i:i+250]))
```

```python
x10 = np.linspace(0, len(close), len(close)-9)
x20 = np.linspace(0, len(close), len(close)-19)
x60 = np.linspace(0, len(close), len(close)-59)
x120 = np.linspace(0, len(close), len(close)-119)
x250 = np.linspace(0, len(close), len(close)-249)
```

```python
plt.plot(close, 'r')
plt.plot(x10, ma10, 'y')
plt.plot(x20, ma20, 'b')
plt.plot(x60, ma60, 'k')
plt.plot(x120, ma120, 'g')
plt.plot(x250, ma250, 'm')
```

## 解法二：用rolling

```python
df['MA10'] = pd.Series(df.iloc[::-1]['Adj Close']).rolling(10).mean()
df['MA20'] = pd.Series(df.iloc[::-1]['Adj Close']).rolling(20).mean()
df['MA60'] = pd.Series(df.iloc[::-1]['Adj Close']).rolling(60).mean()
df['MA120'] = pd.Series(df.iloc[::-1]['Adj Close']).rolling(120).mean()
df['MA250'] = pd.Series(df.iloc[::-1]['Adj Close']).rolling(250).mean()
```

```python
plt.plot(list(reversed(df['Adj Close'])), 'r')
plt.plot(list(reversed(df.MA10)), 'y')
plt.plot(list(reversed(df.MA20)), 'b')
plt.plot(list(reversed(df.MA60)), 'k')
plt.plot(list(reversed(df.MA120)), 'g')
plt.plot(list(reversed(df.MA250)), 'm')
```
