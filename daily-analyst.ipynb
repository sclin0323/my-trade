## 追蹤-個股權重-日

```python
import pandas as pd
import numpy as np

stocks = []
with open('files/stocks/追蹤_個股權值_日_2330.TW.log', 'r', encoding = 'ISO-8859-1') as f:
    for line in f:
        array = line.split(' ')
        o = {}
        o['日期'] = array[0]
        o['股票'] = array[1].split('.')[0]
        o['月線'] = '月上' if array[2] == '1.000000' else '月下'
        o['季線'] = '季上' if array[3] == '1.000000' else '季下'
        o['DIF'] = 'DIF上' if array[4] == '1.000000' else 'DIF下'
        o['Open'] = array[5]
        o['Close'] = array[6]
        o['High'] = array[7]
        o['Low'] = array[8]
        o['成交量'] = array[9]
        o['外資買進'] = array[10]
        o['外資賣出'] = array[11]
        stocks.append(o)
        
df = pd.DataFrame(stocks)

df['Open'] = df['Open'].astype(float)
df['Close'] = df['Close'].astype(float)
df['High'] = df['High'].astype(float)
df['Low'] = df['Low'].astype(float)
df['外資買進'] = df['外資買進'].astype(float)
df['外資賣出'] = df['外資賣出'].astype(float)
df['外資買賣差'] = df['外資買進'] - df['外資賣出']

df[['日期','股票','月線','季線','DIF','Open','Close','High','Low','成交量','外資買進','外資賣出','外資買賣差']]


#df = pd.read_csv('files/stocks/追蹤_個股權值_日_2317.TW.log')
```
